```python
import requests
from bs4 import BeautifulSoup
# URL of the PG hostel listing website
url = 'https://example.com/pg-hostels-listing'
# Send a GET request to the URL
response = requests.get(url)
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
# Find all the hostel entries
hostels = soup.find_all('div', class_='hostel-entry')
# Create a list to store hostel data
hostel_data = []
# Loop through each hostel entry and extract relevant information
for hostel in hostels:
    name = hostel.find('h2', class_='hostel-name').text
    twin_sharing_price = hostel.find('span', class_='twin-sharing-price').text
    location = hostel.find('span', class_='location').text
    amenities = hostel.find('div', class_='amenities').text
   # Store the data in a dictionary
    hostel_info = {
        'Name': name,
        'Twin Sharing Price': twin_sharing_price,
        'Location': location,
        'Amenities': amenities
    }
# Append the dictionary to the list
    hostel_data.append(hostel_info)
# Display the scraped data
for hostel in hostel_data:
    print(hostel)
# Optionally, save the data to a CSV file
import pandas as pd
df = pd.DataFrame(hostel_data)
df.to_csv('pg_hostels_data.csv', index=False)
print("Scraping completed! Data saved to 'pg_hostels_data.csv")
df
```

```python
# Correcting Location Features
# Remove the word 'in ' from the Location column
df['Location'] = df['Location'].str.replace('in ', '')
```

DEFINE A FUNCTION TO EXTRACT PRICES,ACCOUNTING FOR OPTIONAL " WITH AC " TEXT.

```python
def extract_price(room_info, room_type):
    pattern = re.compile(rf'{room_type}(?: With AC)? #([\d,]+)')
    match = pattern.search(room_info)
    return int(match.group(1).replace(',', '')) if match else None
# Define a function to check for AC in room info
def check_for_ac(room_info):
    return 'Yes' if 'With AC' in room_info else 'No'
# Apply the function to create new columns for prices
df['Single Room'] = df['Room Info'].apply(lambda x: extract_price(x, 'Single Room'))
df['Twin Sharing'] = df['Room Info'].apply(lambda x: extract_price(x, 'Twin Sharing'))
df['Triple Sharing'] = df['Room Info'].apply(lambda x: extract_price(x, 'Triple Sharing'))
df['Four Sharing'] = df['Room Info'].apply(lambda x: extract_price(x, 'Four Sharing'))
# Apply the function to create a new column for AC information
df['Has AC'] = df['Room Info'].apply(check_for_ac)
```

FUNCTION TO EXTRACT METRO AND METRO DISTANCE.

```python
def extract_metro_info(places_info):
    metro_match = re.search(r'(\d+.\d+ Km from [\w\s]+ Metro Station)', places_info)
    if metro_match:
        return metro_match.group(1).split(' from ')[1], metro_match.group(1).split(' Km ')[0] + ' Km'
    else:
        return None, None
```

```python
# Create the 'Nearest Metro' and 'Metro Distance' columns
df[['Nearest Metro', 'Metro Distance']] = df['Places'].apply(lambda x: pd.Series(extract_metro_info(x)))
```

FUNCTION TO FILTER TYPE COLUMN AND CREATE GENDER COLUMN.

```python
def filter_type_and_extract_gender(value):
    if 'Boys' in value:
        return 'Boys'
    elif 'Girls' in value:
        return 'Girls'
    elif 'Coed' in value:
        return 'Coed'
    else:
        return None
```

```python
 # Create the 'Gender' column and filter the Type column
df['Gender'] = df['Type'].apply(filter_type_and_extract_gender)
df['Type'] = df['Gender']
```

```python
# Drop the specified columns
df.drop(columns=[ 'Price', 'Type', 'Places','Room Info'], inplace=True)
```

Multivariate analysis examines the relationships between multiple variables simultaneously. Here’s a Python script using seaborn and matplotlib to perform multivariate analysis on your dataset.

1.PAIR PLOT (RELATIONSHIP BETWEEN MULTIPLE FEAUTURES):

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/Cleaned_Data.csv")

# Selecting relevant numerical columns for pairplot
numeric_columns = ["Single Room", "Twin Sharing", "Triple Sharing", "Four Sharing", "Metro Distance"]

# Pairplot for numerical variables
sns.pairplot(df[numeric_columns], diag_kind="kde")
plt.show()
```
